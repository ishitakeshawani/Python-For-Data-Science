```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sn
import numpy as np
```

```python
l = pd.read_csv('Airbnb_NYC_2019.csv')
l
```

```python
sn.countplot(x = 'neighbourhood_group',data = l)
plt.show()
```

```python
sn.barplot(x = 'neighbourhood_group',y = 'price',data = l)
plt.show()
```

```python
sn.barplot(x = 'neighbourhood_group',y = 'price',data = l,ci = False)
plt.show()
```

```python
plt.hist(l['price'])
plt.xlabel('price (in us dolars)')
plt.show()
```

```python
plt.hist(l['price'],bins = np.arange(0,1100,40))
plt.xlabel('price (in US dolars)')
plt.show()
```

```python
plt.scatter(x= l['price'],y=l['number_of_reviews'])
plt.xlabel('price')
plt.ylabel('number of reviews')
plt.show()
```

```python
plt.scatter(x= l['price'],y=l['number_of_reviews'])
plt.xlabel('price')
plt.ylabel('number of reviews')
plt.xlim(0,1100)
plt.show()
```

```python
plt.scatter(x= l['price'],y=l['number_of_reviews'],s=5)
plt.xlabel('price')
plt.ylabel('number of reviews')
plt.xlim(0,1100)
plt.show()
```
